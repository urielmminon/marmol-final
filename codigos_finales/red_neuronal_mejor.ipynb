```python
import feather
import numpy as np
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
walmart_df = feather.read_dataframe('Datos/base_nuevas_vars/walmart.feather')
nrow=walmart_df['TripType'].count()
walmart_df_Y=walmart_df['TripType']
walmart_df=walmart_df.drop(['VisitNumber', 'TripType'], axis=1)
```

```python
print(walmart_df)
```

```python
list(walmart_df.columns.values)
```

```python
pwd
```

```python
df_Weekday = pd.get_dummies(walmart_df['Weekday'])
```

```python
walmart_df = pd.concat([walmart_df, df_Weekday], axis=1)
walmart_df = walmart_df.drop('Weekday', axis=1)
```

```python
print(walmart_df)
```

```python
from sklearn import neural_network
```

```python
hyper_param_grid_nn = {'solver': ['lbfgs'], 'max_iter': [1000], 'alpha': [10000], 'hidden_layer_sizes':[30], 'random_state':[5]}
clf_grid = GridSearchCV(neural_network.MLPClassifier(), hyper_param_grid_nn, n_jobs=-1, cv = 4, verbose = 3 , scoring='log_loss')
```

```python
clf_grid.fit(walmart_df, walmart_df_Y)
```

```python
with open("NN_f_mas.pkl", "wb") as f:
     pickle.dump(clf_grid, f)
```

```python
with open("NN_f_mas.pkl", "rb") as f:
     model_loaded = pickle.load(f)
```
